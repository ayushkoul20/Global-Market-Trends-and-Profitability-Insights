```python
#GROUP SALES BY MONTHS
sales_by_month = sales.groupby('month_year', as_index=False)['sales'].sum()
```

```python
#VISUALIZE SALES TRAND BY MONTHS
plt.figure(figsize=(15,6))
plt.plot(sales_by_month['month_year'],sales_by_month['sales'])
plt.xticks(rotation='vertical',size=8)
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.bar(sales_by_month['month_year'],sales_by_month['sales'])
plt.xticks(rotation='vertical',size=8)
plt.show()
```

```python
#DISPLAY MOST SELLING PRODCUTS
# Ensure 'sales' is numeric (just in case)
sales['sales'] = pd.to_numeric(sales['sales'], errors='coerce')

# Group by product_name and sum only 'sales'
products_sales = sales.groupby('product_name', as_index=False)['sales'].sum()

# Sort by sales in descending order
products_sales = products_sales.sort_values('sales', ascending=False)
```

```python
#TOP 10 MOST SALES PRODUCTS
products_sales[:10]
```

```python
# Make sure 'quantity' is numeric
sales['quantity'] = pd.to_numeric(sales['quantity'], errors='coerce')

# Group by product name and sum only quantity
products_by_quantity = sales.groupby('product_name', as_index=False)['quantity'].sum()

# Sort by quantity sold in descending order
products_by_quantity_sorted = products_by_quantity.sort_values('quantity', ascending=False)
```

```python
#TOP 10 MOST QUANTITY SELLING PRODUCTS ITEMS
products_by_quantity_sorted[:10]
```

```python
#VISUALIZE MOST USED SHIP MODS
plt.figure(figsize=(10, 8.5))
sns.countplot(x='ship_mode', data=sales)
plt.title('Most Used Shipping Modes')
plt.xlabel('Shipping Mode')
plt.ylabel('Count')
plt.show()
```

```python
#Distribution of Shiping Modes
ship_mode = pd.DataFrame(sales['ship_mode'].value_counts()).reset_index()

ship_mode = ship_mode.rename({'index': 'shiping_mode', 'ship_mode': 'count'}, axis=1)

ship_mode.head()
```

```python
# Group by ship_mode and count number of orders per mode
ship_mode = sales['ship_mode'].value_counts().reset_index()
ship_mode.columns = ['ship_mode', 'count']  # Rename columns for clarity

# Plot pie chart
fig = plt.figure(figsize=(10, 7))
plt.pie(ship_mode['count'], labels=ship_mode['ship_mode'], autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Orders by Shipping Mode")
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle
plt.tight_layout()
plt.show()
```

```python

# MOST PROFITABLE PRODUCT CATEGORIES & SUB CATEGORIES

# Group by 'category' and 'sub_category', then sum only the 'profit' column
category_by_profit = sales.groupby(['category', 'sub_category'])['profit'].sum().reset_index()

# Sort the result by 'profit' in descending order
category_by_profit = category_by_profit.sort_values('profit', ascending=False)

# Show the result
print(category_by_profit)
```
